```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi

a = np.linspace(0.01, 0.02)
k = list(range(int(np.ceil(100 * 0.01 / pi)), int(np.floor(500 * 0.02 / pi)) + 1))

for el in k:
    plt.plot(a, (el * pi) / a, label=f'k = {el}')
    
plt.grid(color='lavender', linestyle='--')
plt.xlabel("значения a")
plt.ylabel("значения х")
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", borderaxespad=0)

plt.show()
```
