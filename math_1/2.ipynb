```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi

x = np.arange(0, 2*pi, 0.05)
y1 = np.cos(x)
y2 = np.cos(2 * x)

plt.plot(x, y1, color='r', label='cos(x)')
plt.plot(x, y2, color='g', label='cos(-1 * x)')

plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title('Cos Graphs')

plt.legend
plt.show()
```
