```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi

# Эллипс
u = 1.
v = 0.5
a = 2.
b = 1.5

ellipse = np.linspace(0, 2 * pi, 100)
plt.plot(u + a * np.cos(ellipse), v + b * np.sin(ellipse), color='c')

# Окружность
circle = plt.Circle((5, -2), 1, color='purple', fill=False)
ax = plt.gca()
ax.add_patch(circle)

# Гипербола
x = np.arange(-10, 10)
y = 4 / x + 1  # как добавить здесь исключение что x != 0 ?? вообще питон не падает с ошибкой, а пишет только
# RuntimeWarning: divide by zero encountered in true_divide y = 4 / x + 1
plt.plot(x, y, color='orange')
plt.axis('scaled')

plt.grid(color='lavender', linestyle='--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('Graphs')

plt.show()
```
