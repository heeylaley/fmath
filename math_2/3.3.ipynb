```python
import numpy as np

pol_x = [10,12,10]
pol_y = [0,45,90]

decar = []
for i, j in zip(pol_x, pol_y):
    r = np.sqrt(i ** 2 + j ** 2)
    decar.append((round(r * np.cos(j), 2), round(r * np.sin(j), 2)))

print(decar)
```

```python
import matplotlib.pyplot as plt
pi = np.pi

rads = np.arange(0, (2 * np.pi), 0.01)

for rad in rads:
    plt.polar(rad, 2, 'm.')

plt.show()
```
