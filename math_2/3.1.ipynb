```python
import matplotlib.pyplot as plt
import numpy as np
pi = np.pi

plt.figure(figsize=(8,8))
x = np.linspace(-4 * pi, 4 * pi, 200)
plt.plot(x, 3 * np.cos(x + 2), color='mediumorchid', label='3*cos(x+2)')
plt.plot(x, 0.5 * np.cos(3 * x), color='salmon', label='0.5*cos(3x)')
plt.plot(x, -2 * np.cos(-0.5 * x + 0.5), color='steelblue', label='-2*cos(-0.5x+0.5)')

plt.grid(color='lightgray', linestyle='--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=3, mode="expand", borderaxespad=0)
plt.title('Cos Graphs')

plt.show()
```
