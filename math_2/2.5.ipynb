```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

x, y = np.linspace(-8, 8, 100), np.linspace(-8, 8, 100)
X, Y = np.meshgrid(x, y)
Z1 = 11 - X + Y
Z2 = 20 - X + Y

ax.plot_surface(X, Y, Z1, alpha=0.5, facecolors='c', rstride=100, cstride=100)
ax.plot_surface(X, Y, Z2, alpha=0.5, facecolors='m', rstride=100, cstride=100)

plt.show()
```

```python
from matplotlib import cm

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)

X, Y = np.meshgrid(x, y)
z = X ** 2 - Y ** 2
Z = z

surf = ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap=cm.coolwarm)

plt.show()
```
