```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

x = np.linspace(-2, 3, 201)
plt.plot(x, x ** 2 - 1, color='orange', label='y = x^2 - 1')
plt.plot(x, (np.exp(x) - 1)/x + 1, color='c', label='y = exp(x) - 1)/x + 1')
plt.ylim(-2, 8)

plt.grid(color='lavender', linestyle='--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('Graphs')
plt.legend(bbox_to_anchor=(0, -0.15, 1, 0), loc=2, ncol=2, mode="expand", borderaxespad=0)

plt.show()

def equations(p):
    x, y = p
    return x**2 - 1-y, (np.exp(x) - 1)/x +1-y

x1, y1 =  fsolve(equations, (-1, -1))
x2, y2 =  fsolve(equations, (1, 1))
              
print (x1, y1)
print (x2, y2)
```
